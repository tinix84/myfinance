## Example of APPLE analysis
for example if we want to have a look at APPLE we get all the data from 2000-2019

```python
data = df_annual_balance.drop(df_annual_balance[df_annual_balance['Ticker'] != 'AAPL'].index)
data
```

from the table above we can plot many different graph

```python
# Create figure with secondary y-axis
fig = make_subplots()

# Add traces
fig.add_trace(
    go.Scatter(x=data['Fiscal Year'], y=data['Cash, Cash Equivalents & Short Term Investments'], 
               name="Cash"), secondary_y=False,)

fig.add_trace(
    go.Scatter(x=data['Fiscal Year'], y=data['Retained Earnings'], 
               name="Earnings"), secondary_y=False,)

fig.add_trace(
    go.Scatter(x=data['Fiscal Year'], y=data['Total Liabilities'], 
               name="Total Liabilities"), secondary_y=False,)

# Add figure title
fig.update_layout(title_text="AAPL")

# Set x-axis title
fig.update_xaxes(title_text="Fiscal Year")

# Set y-axes titles
fig.update_yaxes(title_text="Prim Axes [USD]", secondary_y=False)
# fig.update_yaxes(title_text="Retained Earnings", secondary_y=True)

fig.show()
```

```python
fig = make_subplots(
    rows=1, cols=3,
    subplot_titles=("2017","2018", "2019"))

data = df_annual_balance.drop(df_annual_balance[df_annual_balance['Fiscal Year'] != 2017].index)
fig.add_trace(go.Scatter(x=data['Cash, Cash Equivalents & Short Term Investments'], y=data['Retained Earnings'],
                                mode='markers', marker_color=data['SimFinId'],
                                text=data['Ticker']), # hover text goes here
                                row=1, col=1)

data = df_annual_balance.drop(df_annual_balance[df_annual_balance['Fiscal Year'] != 2018].index)
fig.add_trace(go.Scatter(x=data['Cash, Cash Equivalents & Short Term Investments'], y=data['Retained Earnings'],
                                mode='markers', marker_color=data['SimFinId'],
                                text=data['Ticker']), # hover text goes here
                                row=1, col=2)

data = df_annual_balance.drop(df_annual_balance[df_annual_balance['Fiscal Year'] != 2019].index)
fig.add_trace(go.Scatter(x=data['Cash, Cash Equivalents & Short Term Investments'], y=data['Retained Earnings'],
                                mode='markers', marker_color=data['SimFinId'],
                                text=data['Ticker']), # hover text goes here
                                row=1, col=3)


fig.update_layout(showlegend=False, title_text="Retained Earnings vs Cash over years")
fig.show()
```
